```python
# %% [markdown]
## 1) Imports & device
import os, io, math, random, time, zipfile, tarfile, json, shutil
from pathlib import Path
from typing import Tuple

import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader

import torchvision
from torchvision import transforms
from torchvision.datasets import ImageFolder
from torchvision.models import resnet18, ResNet18_Weights

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from tqdm import tqdm

plt.rcParams['figure.dpi'] = 120
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
DEVICE
```

```python
# %% [markdown]
## 2) Get data: auto‑download PRPD images (Mendeley) or synthesize fallback
from urllib.request import urlopen, Request
import urllib.error

DATA_DIR = Path('data')
DATASET_NAME = 'mendeley_prpd'  # or 'synth_prpd' on fallback
ROOT = DATA_DIR / DATASET_NAME

def attempt_mendeley_download(dest: Path) -> bool:
    """Try to download a small public set of PRPD images from Mendeley.
    We use the DOI page: https://data.mendeley.com/datasets/xz4xhrc4yr/7
    NOTE: If the portal enforces cookies/auth or rate limits, this will fail gracefully.
    """
    # Some Mendeley pages expose direct file endpoints under /public-files/datasets/.../files/<fileid>?download=1
    # These IDs can change with new versions; we include a minimal curated list that worked previously,
    # but we also guard for failures and fallback to synthetic data.
    files = [
        # (filename, public-file URL)
        # If these links stop working, the code will fallback automatically.
        # Example SMALL sample zip made by authors (if available). Otherwise, we try the page-level 'download all'.
        ('download_all.zip', 'https://prod-dcd-datasets-cache-zipfiles.s3.eu-west-1.amazonaws.com/xz4xhrc4yr-7.zip'),
    ]

    dest.mkdir(parents=True, exist_ok=True)
    ok = False
    for fname, url in files:
        try:
            print(f'Trying {url}')
            req = Request(url, headers={'User-Agent':'Mozilla/5.0'})
            with urlopen(req, timeout=60) as r:
                data = r.read()
            arc = dest / fname
            with open(arc, 'wb') as f:
                f.write(data)
            # Try unzip
            try:
                with zipfile.ZipFile(arc) as zf:
                    zf.extractall(dest)
                    ok = True
            except zipfile.BadZipFile:
                pass
            # Try tar if not zip
            if not ok:
                try:
                    with tarfile.open(arc) as tf:
                        tf.extractall(dest)
                        ok = True
                except tarfile.ReadError:
                    pass
            if ok:
                print('Mendeley PRPD dataset extracted to', dest)
                break
        except Exception as e:
            print('Download attempt failed:', e)
    # Heuristic: expect class folders like corona/surface/internal
    if ok:
        subdirs = [p for p in dest.glob('*') if p.is_dir()]
        if not subdirs:
            ok = False
    return ok

def synthesize_prpd(dest: Path, n_per_class: int = 400, img_size: int = 224, seed: int = 0):
    """Generate a toy yet plausible PRPD image dataset with three classes: corona, surface, internal.
    Each image is a heat-like 2D map constructed from Gaussian blobs modulated across phase bins.
    """
    import numpy as np
    import matplotlib.pyplot as plt
    rng = np.random.default_rng(seed)
    dest.mkdir(parents=True, exist_ok=True)
    classes = ['corona', 'surface', 'internal']
    for c in classes:
        (dest / c).mkdir(parents=True, exist_ok=True)

    H = W = img_size
    phase = np.linspace(0, 2*np.pi, W)

    def blob(img, cx, cy, sx, sy, amp):
        yy, xx = np.mgrid[0:H, 0:W]
        img += amp * np.exp(-((xx-cx)**2/(2*sx**2) + (yy-cy)**2/(2*sy**2)))
        return img

    for cls in classes:
        for i in range(n_per_class):
            img = np.zeros((H, W), dtype=np.float32)
            # Two half-cycles with class‑specific shape cues
            # corona: triangular-ish bands around peaks
            # surface: broader bands near zero-crossings with spread
            # internal: arcs near quarter phases
            for half in [0, 1]:
                base = rng.normal(0.0, 0.02, size=(H, W)).astype(np.float32)
                center_x = (0.25 if cls=='internal' else 0.15 if cls=='surface' else 0.1) * W + half*W*0.5 + rng.normal(0, 2.0)
                spread_x = (0.18 if cls=='internal' else 0.28 if cls=='surface' else 0.12) * W
                # vertical position encodes amplitude clusters
                bands = 2 if cls=='corona' else 3 if cls=='surface' else 2
                for b in range(bands):
                    cy = H*(0.35 + 0.25*b) + rng.normal(0, 3.0)
                    sx = spread_x * (0.7 + 0.6*rng.random())
                    sy = (6 + 18*rng.random())
                    amp = 0.7 + 0.6*rng.random()
                    base = blob(base, center_x + (b-1)*0.15*W*(0.6 if cls=='corona' else 0.9), cy, sx, sy, amp)
                # subtle phase-modulated ripple
                for x in range(W):
                    k = 0.1*np.sin(phase[x]* (2 if cls=='internal' else 3))
                    base[:, x] += k
                img += base
            # Normalize to [0,1]
            img = img - img.min()
            if img.max() > 0: img = img / img.max()
            # Save as PNG RGB
            from PIL import Image
            arr = (img*255).astype(np.uint8)
            rgb = np.stack([arr, arr, arr], axis=-1)
            Image.fromarray(rgb).save(dest / cls / f"{cls}_{i:05d}.png")
    print('Synthetic PRPD dataset created at', dest)

# Try download; else synthesize
ok = attempt_mendeley_download(ROOT)
if not ok:
    print('Falling back to synthetic PRPD dataset...')
    DATASET_NAME = 'synth_prpd'
    ROOT = DATA_DIR / DATASET_NAME
    synthesize_prpd(ROOT, n_per_class=400, img_size=224, seed=0)

ROOT, DATASET_NAME
```

```python
# %% [markdown]
## 3) Data loading, transforms, stratified split
IMG_SIZE = 128
BATCH_SIZE = 32
VAL_SPLIT = 0.15
TEST_SPLIT = 0.15
SEED = 42

torch.manual_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
])

full_ds = ImageFolder(ROOT, transform=transform)
labels = [y for _, y in full_ds.imgs]
if len(labels) == 0:
    raise RuntimeError(f'No images found in {ROOT}.')

idx = np.arange(len(labels))
idx_train, idx_tmp, y_train, y_tmp = train_test_split(idx, labels, test_size=VAL_SPLIT+TEST_SPLIT, stratify=labels, random_state=SEED)
rel = TEST_SPLIT/(VAL_SPLIT+TEST_SPLIT)
idx_val, idx_test, y_val, y_test = train_test_split(idx_tmp, y_tmp, test_size=rel, stratify=y_tmp, random_state=SEED)

class Subset(torch.utils.data.Dataset):
    def __init__(self, base, indices):
        self.base, self.indices = base, indices
        self.samples = [base.samples[i] for i in indices]
        self.targets = [base.targets[i] for i in indices]
        self.classes = base.classes
        self.class_to_idx = base.class_to_idx
        self.transform = base.transform
    def __len__(self): return len(self.indices)
    def __getitem__(self, i): return self.base[self.indices[i]]

train_ds = Subset(full_ds, idx_train)
val_ds   = Subset(full_ds, idx_val)
test_ds  = Subset(full_ds, idx_test)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

len(train_ds), len(val_ds), len(test_ds), full_ds.classes
```

```python
# %% [markdown]
## 4) Peek at the data
def show_batch(dl, n=12):
    imgs, labs = next(iter(dl))
    n = min(n, imgs.size(0))
    cols = 6; rows = math.ceil(n/cols)
    plt.figure(figsize=(cols*2, rows*2))
    for i in range(n):
        plt.subplot(rows, cols, i+1)
        plt.imshow(np.transpose(imgs[i].numpy(), (1,2,0)))
        plt.title(full_ds.classes[labs[i]])
        plt.axis('off')
    plt.tight_layout(); plt.show()

show_batch(train_loader, n=12)
```

```python
# %% [markdown]
## 5) Model (ResNet18) & training
num_classes = len(full_ds.classes)
weights = ResNet18_Weights.DEFAULT
model = resnet18(weights=weights)
model.fc = nn.Linear(model.fc.in_features, num_classes)
model = model.to(DEVICE)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4, weight_decay=1e-4)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', patience=2, factor=0.5)

def accuracy(pred, target):
    return (pred.argmax(dim=1) == target).float().mean().item()

history = {'epoch':[], 'train_loss':[], 'train_acc':[], 'val_loss':[], 'val_acc':[]}
EPOCHS = 3
for epoch in range(1, EPOCHS+1):
    model.train(); tl, ta = 0.0, 0.0
    for xb, yb in tqdm(train_loader, desc=f'Epoch {epoch}/{EPOCHS} [train]'):
        xb, yb = xb.to(DEVICE), yb.to(DEVICE)
        optimizer.zero_grad()
        out = model(xb)
        loss = criterion(out, yb)
        loss.backward(); optimizer.step()
        tl += loss.item() * xb.size(0)
        ta += accuracy(out.detach().cpu(), yb.detach().cpu()) * xb.size(0)
    tl /= len(train_ds); ta /= len(train_ds)

    model.eval(); vl, va = 0.0, 0.0
    with torch.no_grad():
        for xb, yb in tqdm(val_loader, desc=f'Epoch {epoch}/{EPOCHS} [val]'):
            xb, yb = xb.to(DEVICE), yb.to(DEVICE)
            out = model(xb)
            loss = criterion(out, yb)
            vl += loss.item() * xb.size(0)
            va += accuracy(out.cpu(), yb.cpu()) * xb.size(0)
    vl /= len(val_ds); va /= len(val_ds)
    scheduler.step(va)

    history['epoch'].append(epoch)
    history['train_loss'].append(tl)
    history['train_acc'].append(ta)
    history['val_loss'].append(vl)
    history['val_acc'].append(va)
    print(f'Epoch {epoch}: train_acc={ta:.3f}, val_acc={va:.3f}')

plt.figure(); plt.plot(history['epoch'], history['train_loss'], label='train_loss'); plt.plot(history['epoch'], history['val_loss'], label='val_loss'); plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.title('Loss Curves'); plt.legend(); plt.show()
plt.figure(); plt.plot(history['epoch'], history['train_acc'], label='train_acc'); plt.plot(history['epoch'], history['val_acc'], label='val_acc'); plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.title('Accuracy Curves'); plt.legend(); plt.show()
```

```python
# %% [markdown]
## 6) Evaluation — confusion matrix & report
model.eval()
all_preds, all_labels = [], []
with torch.no_grad():
    for xb, yb in test_loader:
        xb = xb.to(DEVICE)
        out = model(xb).cpu()
        preds = out.argmax(dim=1).numpy()
        all_preds.extend(list(preds))
        all_labels.extend(list(yb.numpy()))

cm = confusion_matrix(all_labels, all_preds)
classes = full_ds.classes

plt.figure(figsize=(6,5))
plt.imshow(cm, interpolation='nearest')
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes, rotation=45, ha='right')
plt.yticks(tick_marks, classes)
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, str(cm[i, j]), ha='center', va='center')
plt.ylabel('True label'); plt.xlabel('Predicted label')
plt.tight_layout(); plt.show()

print(classification_report(all_labels, all_preds, target_names=classes))
```

```python
# %% [markdown]
## 7) Qualitative results — prediction grid
def show_predictions(dl, n=12):
    imgs, labs = next(iter(dl))
    model.eval()
    with torch.no_grad():
        out = model(imgs.to(DEVICE)).cpu()
        pred = out.argmax(dim=1)
        prob = torch.softmax(out, dim=1).max(dim=1).values
    n = min(n, imgs.size(0))
    cols = 6; rows = math.ceil(n/cols)
    plt.figure(figsize=(cols*2, rows*2))
    for i in range(n):
        plt.subplot(rows, cols, i+1)
        plt.imshow(np.transpose(imgs[i].numpy(), (1,2,0)))
        ttl = f'pred: {full_ds.classes[int(pred[i])]}\ntrue: {full_ds.classes[int(labs[i])]}\np={float(prob[i]):.2f}'
        plt.title(ttl); plt.axis('off')
    plt.tight_layout(); plt.show()

show_predictions(test_loader, n=12)
```

```python
# %% [markdown]
## 8) (Optional) Grad‑CAM
def gradcam(model, x, target_layer='layer4'):
    model.eval(); feats = {}; grads = {}
    def fwd_hook(_, __, out): feats['v'] = out.detach()
    def bwd_hook(_, grad_input, grad_output): grads['v'] = grad_output[0].detach()
    layer = dict(model.named_modules())[target_layer]
    h1 = layer.register_forward_hook(fwd_hook)
    h2 = layer.register_full_backward_hook(bwd_hook)
    out = model(x)
    cls = out.argmax(dim=1)
    score = out.gather(1, cls.view(-1,1)).squeeze()
    model.zero_grad(); score.backward(torch.ones_like(score))
    g = grads['v']; a = g.mean(dim=(2,3), keepdim=True)
    f = feats['v']; cam = (a * f).sum(dim=1, keepdim=True)
    cam = torch.relu(cam); cam = (cam - cam.min()) / (cam.max() - cam.min() + 1e-7)
    h1.remove(); h2.remove(); return cam, cls

imgs, labs = next(iter(test_loader))
xb = imgs.to(DEVICE)[:8]
cam, cls = gradcam(model, xb)
cam = cam.cpu().numpy()
cols = 4; rows = math.ceil(xb.size(0)/cols)
plt.figure(figsize=(cols*3, rows*3))
for i in range(xb.size(0)):
    plt.subplot(rows, cols, i+1)
    base = np.transpose(imgs[i].numpy(), (1,2,0))
    heat = cam[i,0]
    heat = (heat - heat.min()) / (heat.max() - heat.min() + 1e-7)
    blend = 0.4*base + 0.6*heat[...,None]
    plt.imshow(blend)
    plt.title(f'pred={full_ds.classes[int(cls[i])]}\ntrue={full_ds.classes[int(labs[i])]}')
    plt.axis('off')
plt.tight_layout(); plt.show()
```
